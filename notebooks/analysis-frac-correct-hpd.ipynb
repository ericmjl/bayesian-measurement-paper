```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle as pkl

sns.set_context('poster')
sns.set_style('white')
```

```python
!ls sim_results/
```

```python
with open('sim_results/max_21_reps_20_sims.pkl', 'rb') as f:
    n_reps, frac_correct = pkl.load(f)
```

```python
data = pd.DataFrame()
data['n_reps'] = n_reps
data['frac_correct'] = frac_correct
data.head()
```

```python
sns.boxplot(x='n_reps', y="frac_correct", data=data)
plt.xlabel('Num. Replicate Measurements')
plt.ylabel('Fraction Actual in HPD')
plt.savefig('../figures/accuracy.pdf', bbox_inches='tight', )
```
